```python
from volleyball.conf import csv_explorer
import pandas
import plotly.express as plotly_express
from matplotlib import pyplot
```

```python
csv_file = csv_explorer('world_grand_prix-final.csv', extra_paths=['consolidated'])
data = pandas.read_csv(csv_file)
```

```python
df = pandas.DataFrame(data=data, columns=['name','date_of_birth','height','weight','spike','block'])
```

```python
df.head()
```

```python
figure = plotly_express.scatter_matrix(df, dimensions=['height', 'weight', 'spike', 'block'], title='Correlation between charateristics', width=800, height=800)
figure.show()
```
